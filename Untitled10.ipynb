```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:/Users/gagan/Downloads/mtcars.csv")

# Display the first few rows to understand the dataset
print(df.head())

# Choose the variable to analyze (e.g., 'mpg')
variable = 'mpg'

# Create histogram
counts, bins, patches = plt.hist(df[variable], bins=10, edgecolor='black')
plt.xlabel(variable)
plt.ylabel('Frequency')
plt.title(f'Histogram of {variable}')
plt.show()

# Find the interval with the highest frequency
max_count_index = counts.argmax()
highest_frequency_interval = (bins[max_count_index], bins[max_count_index + 1])

print(f'Interval with highest frequency: {highest_frequency_interval}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:/Users/gagan/Downloads/mtcars.csv")

# Display the first few rows to understand the dataset
print(df.head())

# Verify column names
print(df.columns)

# Count frequencies of transmission types
transmission_counts = df['am'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 5))
transmission_counts.plot(kind='bar', color=['blue', 'orange'], edgecolor='black')
plt.xlabel('Transmission Type')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Transmission Types')
plt.xticks(ticks=[0, 1], labels=['Automatic', 'Manual'], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:/Users/gagan/Downloads/mtcars.csv")

# Create a box plot for the 'mpg' column
plt.figure(figsize=(8, 6))
plt.boxplot(df['mpg'], vert=False, patch_artist=True)
plt.xlabel('Miles per Gallon (mpg)')
plt.title('Box Plot of Miles per Gallon (mpg)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Compute the five-number summary
five_num_summary = df['mpg'].describe(percentiles=[.25, .5, .75])
print(five_num_summary)
```
